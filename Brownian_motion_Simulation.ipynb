```python
import numpy as np
import matplotlib.pyplot as plt
n = 2000
dt = 1/n
W0 = 0
BM = []
for i in range(1):
    z =np.random.normal(0,1,n)
    dWt = np.sqrt(dt)*z
    inc = np.cumsum(dWt)
    W = np.append(W0,inc)
    BM.append(W)
    
X = np.array(BM)
t = np.linspace(0, 1, num=n+1)
plt.plot(t,X.T)
plt.title('Brownian Motion Simulation')
plt.show()
    
```
